```python
import pandas as pd
```

```python
df = pd.read_csv("Earthquakes_database.csv")
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
df['Date'] = df['Date'] + ' ' + df['Time']
df = df.drop(columns=['Time'])
```

```python
df=df.fillna('Information Not Available')
df.head()
```

```python
magnitude_bins = [0, 5, 7, float('inf')]  # Adjust the bin edges based on your requirements
magnitude_labels = ['Low Magnitude', 'Medium Magnitude', 'High Magnitude']
df['Magnitude Category'] = pd.cut(df['Magnitude'], bins=magnitude_bins, labels=magnitude_labels, right=False)
df.head()
```

```python
subset_columns = ['Date', 'Magnitude Category', 'Magnitude']
melted_df = pd.melt(df[subset_columns], id_vars=['Date', 'Magnitude Category'], value_vars='Magnitude',
                    value_name='Magnitude Value')
print("Melted DataFrame:")
print(melted_df.head())
```

```python
pivot_table_df = pd.pivot_table(melted_df, values='Magnitude Value', index=['Date'], columns=['Magnitude Category'],
                                aggfunc='count', fill_value=0)

print("\nReshaped DataFrame:")
print(pivot_table_df.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Line Chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='Magnitude Type', y='Magnitude', data=df)
plt.title('Line Chart')
plt.xlabel('Magnitude Type')
plt.ylabel('Magnitude')
plt.show()
```

```python
# 2. Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Magnitude Category', y='Magnitude', data=df, estimator=len)
plt.title('Bar Chart - Count of Earthquakes by Magnitude Category')
plt.xlabel('Magnitude Category')
plt.ylabel('Count of Earthquakes')
plt.show()
```

```python
# 3. Scatter Chart
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Magnitude', y='Depth', data=df, alpha=0.5)
plt.title('Scatter Chart - Relationship between Magnitude and Depth')
plt.xlabel('Magnitude')
plt.ylabel('Depth')
plt.show()
```

```python
# 4. Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Magnitude'], bins=20, kde=True)
plt.title('Histogram - Distribution of Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format=True, errors='coerce')

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day

df.head()
```
